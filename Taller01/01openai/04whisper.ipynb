## Transcribiendo texto desde un archivo mp3 con Whisper

```python
!wget https://bigdataiotcloudaws.s3.amazonaws.com/files/cliente-enojado.mp3
```

```python
from openai import OpenAI
import getpass, os
```

```python
os.environ["OPENAI_API_KEY"] = getpass.getpass("Ingresa tu API Key de OpenAI")
```

```python
client = OpenAI()

audio_file= open("cliente-enojado.mp3", "rb")
transcript = client.audio.transcriptions.create(
  model = "whisper-1", 
  file = audio_file
)
```

```python
print(transcript.text)
```
