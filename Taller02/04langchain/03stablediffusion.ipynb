## Generando imágenes con Stable Diffusion

```python
import os, getpass
from PIL import Image
import requests
from io import BytesIO
from langchain_community.llms import Replicate
```

```python
os.environ["REPLICATE_API_TOKEN"] = getpass.getpass("Ingresa tu API Key de Replicate : ")
```

```python
llm_generate_image = Replicate(
    model="stability-ai/sdxl:39ed52f2a78e934b3ba6e2a89f5b1c712de7dfea535525255b1aa35c5565e08b",
    model_kwargs={
        "prompt": "A flying armored superhero next to the Eiffel Tower"
    }
)

#image_output = llm_generate_image("Iron Man flying next to the Eiffel Tower")
#response = requests.get(image_output)
#img = Image.open(BytesIO(response.content))

image_output = llm_generate_image.invoke("Iron Man flying next to the Eiffel Tower")
```

```python
response = requests.get(image_output)
img = Image.open(BytesIO(response.content))
img
```
